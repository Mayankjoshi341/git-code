```python
netflix = pd.read_csv("C:\\Users\\mayan\\OneDrive\\git-projects\\git-code\\MyData-Warehouse\\data-cleaning\\cleaned-csv\\Netflix_dataset.csv")
```

```python
netflix.head()
```

```python
def colmun_info():
    for col in netflix.columns:
        print(f"Column: {col}")
        print(f"Unique Values: {netflix[col].nunique()}")
        if netflix[col].nunique() < 15:
            print(f"Unique Values Sample: {netflix[col].unique()}")
        print(f"Missing Values: {netflix[col].isnull().sum()}")
        print(f"Data Type: {netflix[col].dtype}")
        print("-" * 25)
colmun_info()
```

```python
sns.set(style="darkgrid")
netflix["type"].value_counts().plot(kind='bar', figsize=(4, 4), color=['#ff9999', 'skyblue'], edgecolor='black')
plt.title("Distribution of Content Type" , fontsize=16 , fontweight='bold')
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()

plt.pie(netflix["type"].value_counts(), labels= netflix["type"].unique(), autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title("Content Type Distribution")
plt.figure(figsize=(4, 4))
plt.tight_layout()
plt.show()
```

```python
reating_counts = netflix['rating'].value_counts()
plt.pie(reating_counts[:7], labels=reating_counts.index[:7], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Ratings Distribution in percentage" , fontstyle='italic', fontsize=16)
plt.show()
```

```python
moive_duration =  netflix[netflix["type"] == "Movie"]["duration"].str.replace(" min", "").astype(int)
moive_duration.value_counts().head(10)
```

```python
plt.hist(moive_duration , bins = 30 , color = 'limegreen', edgecolor='black')
plt.title("Distribution of Movie Duration", fontsize=16, fontweight='bold')
plt.xlabel("Duration in minutes")
plt.ylabel("Frequency of Movies")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```

```python
netflix["release_year"].value_counts().sort_index().plot(kind='line', figsize=(12, 6), color='#ff9999', linestyle = "--" , marker='o' , markersize=5)
```

```python
year_by_rating = netflix["rating"].groupby(netflix["release_year"]).value_counts().sort_index().tail(40)
print(type(year_by_rating))
year_by_rating[0:12].plot(kind = "line", figsize=(12, 6), color='skyblue', linestyle='--', marker='o', markersize=5)
#year_by_rating[12:].plot(kind = "line", figsize=(12, 6), color='orange', linestyle='--', marker='^', markersize=5)
plt.title("Ratings by Year", fontsize=16, fontweight='bold')
```
