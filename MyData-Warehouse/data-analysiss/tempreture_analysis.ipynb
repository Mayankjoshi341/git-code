```python
temp_year1 = (temp[temp["YEAR"] == 1901])
temp_year2 = (temp[temp["YEAR"] == 2017])
```

```python
temp_year2
df_melted2 = pd.melt(temp_year2, id_vars = ["YEAR"], value_vars = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN", "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]
                     , var_name = "Months", value_name = "Temperature")
df_melted2
```

```python
columns_to_plot = ["JAN", "FEB", "MAR", "APR", "MAY", "JUN","JUL", "AUG", "SEP", "OCT", "NOV", "DEC"]
df_melted = temp_year1.melt(id_vars = "YEAR", value_vars= columns_to_plot, 
                      var_name= "Months", value_name= "Temperature")
df_melted
```

```python
x = df_melted["Months"]
y1 = df_melted["Temperature"]
y2 = df_melted2["Temperature"]
plt.plot(x, y1, color = "red", marker = "o", linestyle = "dashed", linewidth = 2, markersize = 5)
plt.plot(x , y2 , color = "lime" , marker = 'd' , linestyle = "-", linewidth = 2, markersize = 5)
plt.legend(["1901", "2017"])
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.ylim(20,40)
plt.title("Temperature in 1901 and 2017")
plt.show()
```
