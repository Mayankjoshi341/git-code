<h1> EXPONENTAL ANALYSIS </h1>

```python
plt.figure(figsize=(28 , 8))
x = sns.countplot(data = pizza , x = "total_price")
x.bar_label(x.containers[0])
plt.show()
```

```python
plt.figure(figsize=(15 , 3))
sns.boxplot(data = pizza , x = "total_price")
```

```python
plt.figure(figsize = (28, 7))
x = sns.countplot(data = pizza , x = "total_price" , hue = "pizza_size")
x.bar_label(x.containers[0])
plt.show()
```

```python
x = sns.countplot(data = pizza , x = "pizza_size")
x.bar_label(x.containers[0])
plt.show()
```
